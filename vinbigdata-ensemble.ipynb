```python
subcsv = pd.read_csv('../input/ensemble2classvinbigdata/submissions-mult.csv')
```

```python
ind = 0
for index, row in subcsv.iterrows():
    if sub_2c[ind][1]=='Normal' and row['PredictionString'][:2] != '14':
        if sub_2c[ind][2].item() > 0.6 and sub_2c[ind][2].item() < 0.95:
            row['PredictionString'] += f" 14 {sub_2c[ind][2].item()} 0 0 1 1"
        else:
            row['PredictionString'] = f" 14 1 0 0 1 1" 
    elif sub_2c[ind][1]=='Normal' and sub_2c[ind][2].item() < 0.95:
        row['PredictionString'] = f" 14 {sub_2c[ind][2].item()} 0 0 1 1"
    ind+=1
```
